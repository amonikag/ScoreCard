```
#importing all necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import jinja2
import xlrd
from PIL import Image, ImageDraw, ImageFont
from matplotlib.backends.backend_pdf import PdfPages
import reportlab
from reportlab.pdfgen import canvas
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.units import inch
from reportlab.platypus import Frame,Image,Table,TableStyle,SimpleDocTemplate,Flowable,PageBreak
import os
import glob
```

```
#open excel sheet
import pandas as pd
file = pd.read_excel(r'/content/drive/My Drive/intern_project/Assignment.xlsx')
file.head(4)
```

```
#creating necessary data frames and arrays from excel sheet and appending to code
Name=[]
City=[]
Country=[]
ExtraTime=[]
Num=[]
Grade=[]
Gender=[]
DOB=[]
DOT=[]
SchoolName=[]
Time=[]
l=0
```

```
#modify
cnt=0
for i in range(0,50,5):
  cnt+=1
  # for Name and registration number
  Name.append(file.iloc[i]['Name of Candidate'])
  Num.append(file.iloc[i]['Registration'])
  #For Gender
  if (file.iloc[i]['Gender']=='M'):
    Gender.append('Male')
  elif (file.iloc[i]['Gender']=='F'):
    Gender.append('Female')
  else:
    Gender.append('Others')
    
  #For Grade
  Grade.append(file.iloc[i]['Grade '])

  #For Name of School
  SchoolName.append(file.iloc[i]['Name of school '])

  #For Date of Birth
  year,month,day=str(file.iloc[i]['Date of Birth ']).split('-')
  day=day.split(' ')[0]
  dae='{}-{}-{}'.format(day,month,year)
  DOB.append(dae)

  #For Date of Test
  
  y,m,d=str(file.iloc[i]['Date and time of test']).split('-')
  d=d.split(' ')[0]
  date='{}-{}-{}'.format(d,m,y)
  DOT.append(date)


  #For City of Residence
  City.append(file.iloc[i]['City of Residence'])

  #For Country of Residence
  Country.append(file.iloc[i]['Country of Residence'])

  #For extra time
  Time.append(file.iloc[i]['Extra time assistance'])
file["What you marked"].fillna("-", inplace = True) 
```

```
#uppertitle data
img1='/content/drive/My Drive/intern_project/Logo'
title='STUDENT'
title2='REPORT CARD'
subTitle='WISDOM tests'
upperContent="Student Name: "+ Name[0]
```

```
#registering fonts
pdfmetrics.registerFont(
    TTFont('head','/content/drive/My Drive/intern_project/PlayfairDisplaySC-Regular[1].ttf')
)
pdfmetrics.registerFont(
    TTFont('con','/content/drive/My Drive/intern_project/PlayfairDisplay-Italic.ttf'))
```

```
#below table score calculation:
per=[]
finalScore=[]
j=0
non=0
tscore=[]
for i in range(1,11):
  tscore.clear()
  for k in range(non,non+5):
    if (file['Student No '][j]==i):
      tscore.append(file['Your score'][k])
    j+=1
  non+=5

  finalScore.append(sum(tscore))
#print(finalScore)
final=sorted(finalScore)
for i in range (0,10):
  per.append((len(final[0:final.index(finalScore[i])+1])*100)/len(finalScore))
```

```
# for bar graph
def barchar(x):
  fig=plt.bar(file['Question No.'][x:x+5],file['Time Spent on question (sec)'][x:x+5])
  plt.xlabel('Question No.',fontsize=15)
  plt.ylabel('Time (Sec)',fontsize=15)
  plt.savefig('bar{}.png'.format(x),transparency=True)
  plt.close()
  plt.clf()
```

```
#for pie chart
def timesp(c):
  labels = 'Q1', 'Q2', 'Q3', 'Q4','Q5'
  fig2=plt.pie(file['Time Spent on question (sec)'][c:c+5],autopct='%1.0f%%', shadow=True, startangle=140)
  plt.legend(labels)
  plt.axis('equal')
  plt.savefig('timesp{}.png'.format(c),transparency=True)
  plt.close()
  plt.clf()
```

```
#for piechart 2
def attempt(x):
  znt=[]
  cnt=0
  cnt1=0
  for i in range(x,x+5):
    if (file['Attempt status '][i]=='Attempted'):
      cnt+=1
    else:
      cnt1+=1

  znt.append(cnt)
  znt.append(cnt1)
  labels1='Attempted','Unattempted'
  fig3=plt.pie(znt,autopct='%1.0f%%',shadow=True)
  plt.legend(labels1,loc=2)
  plt.axis('equal')
  plt.savefig('attempt{}.png'.format(x),transparency=True)
  plt.close()
  plt.clf()
```

```
#accuracy
def Accuracy(z):
  accuracy =[]
  ynt=0
  ynt1=0
  for i in range(z,z+5):
    if (file['Outcome (Correct/Incorrect/Not Attempted)'][i]=='Correct'):
      ynt+=1
    elif(file['Outcome (Correct/Incorrect/Not Attempted)'][i]=='Incorrect'):
      ynt1=+1
  accuracy.append(ynt)
  accuracy.append(ynt1)
  labels='Correct','Incorrect'
  cs=['lightblue','red','darkblue']
  fig4=plt.pie(accuracy,autopct='%1.0f%%',shadow=True,colors=cs)
  plt.legend(labels,loc=1)
  plt.savefig('Accuracy{}.png'.format(z),transparency=True)
  accuracy.clear()
  plt.close()
  plt.clf()
```

```
#performance
def perform(z):
  un=0
  ccuracy =[]
  ynt=0
  ynt1=0
  for i in range(z,z+5):
    if (file['Outcome (Correct/Incorrect/Not Attempted)'][i]=='Correct'):
      ynt+=1
    elif(file['Outcome (Correct/Incorrect/Not Attempted)'][i]=='Incorrect'):
      ynt1=+1
  ccuracy.append(ynt)
  ccuracy.append(ynt1)
  un=5-(ynt+ynt1)
  ccuracy.append(un)
  labels='Correct','Incorrect','Unattempted'
  cs=['green','yellow','blue']
  fig4=plt.pie(ccuracy,autopct='%1.0f%%',shadow=True,colors=cs,startangle=140)
  plt.legend(labels)
  plt.savefig('perform{}.png'.format(z),transparency=True)
  plt.close()
  plt.clf()
  ccuracy.clear()
```
